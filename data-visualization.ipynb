# Data Visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from matplotlib.ticker import PercentFormatter
sns.set()
```

# Bar chart

```python
chart_data_cars = pd.read_csv('bar_chart_data.csv')
chart_data_cars
```

```python
plt.figure(figsize=(9, 6))

plt.bar(chart_data_cars['Brand'], chart_data_cars['Cars Listings'], color=['r', 'g', 'b', 'w', "y", "m","c"])

plt.xticks(rotation=45, fontsize=13)

plt.yticks(fontsize=13)

plt.title("Cars Listing by Brand by Donald Programmer", fontsize=16, fontweight="bold")

plt.ylabel("Numbers of  Listing", fontsize=13)

plt.savefig("Used Cars Bar.pdf")

plt.show()
```

# Pie chart 

```python
pie_data_fuel = pd.read_csv('pie_chart_data.csv')
```

```python
pie_data_fuel
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10, 8))

plt.pie(pie_data_fuel['Number of Cars'], labels=pie_data_fuel['Engine Fuel Type'].values, 
       autopct='%.2f%%',
       textprops = {
           'size': 'x-large',
           'fontweight': 'bold',
           'rotation': 30,
           'color': 'w'
       })

plt.legend()
plt.title(" Number of cars by Engine Fuel by Donald Programmer", fontweight="bold", fontsize=16)
plt.show()
```

# STACK CHART

```python
stack_fuel_data = pd.read_csv("stacked_area_chart_data.csv")
```

```python
stack_fuel_data
```

```python
plt.figure(figsize=(12, 6))
sns.set_style('white')
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Gas', 'Petrol', "Diesel"]
plt.stackplot(stack_fuel_data['Year'], stack_fuel_data['Gas'], stack_fuel_data['Petrol'], stack_fuel_data['Diesel'],
             colors=colors,
             edgecolor='none')

plt.xticks(stack_fuel_data['Year'],  rotation=45)
plt.ylabel("Number of Cars", fontsize=12)
plt.legend(labels=labels, loc='upper left')
plt.title("Popularity of engine fuel from 1982 to 2016 By Donald Programmer", fontsize=16, fontweight='bold')
sns.despine()
plt.savefig("engine.png")
plt.show()
```

# Line Chart

```python
data_fte_spx = pd.read_csv('line_chart_data.csv')
```

```python
data_fte_spx
```

```python
data_fte_spx['New_date']  = pd.to_datetime(data_fte_spx["Date"])
data_fte_spx['New_date'] 
```

```python
labels = ['GSPC 500', 'FTSE 100']
plt.figure(figsize=(20, 6))
plt.plot(data_fte_spx['New_date'], data_fte_spx['GSPC500'])
plt.plot(data_fte_spx['New_date'], data_fte_spx['FTSE100'])
plt.title('GSPC500 VS FTSE100 Returns (2000-2010) By Donald Programmer', fontsize=14, fontweight='bold')
plt.legend(labels=labels, fontsize='large')
plt.xlabel("Date")
plt.ylabel("Return")

plt.savefig('sspc=ftse.png')

plt.show()
```

# Histogram Chart

# Scatter Plot 

```python
data_real_estate_scatter = pd.read_csv('scatter_data.csv')
```

```python
data_real_estate_scatter
```

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(data_real_estate_scatter['Area (ft.)'], 
            data_real_estate_scatter['Price'], 
           alpha=0.6,
            c=data_real_estate_scatter['Building Type'],
           cmap='viridis')
plt.legend(*scatter.legend_elements(), loc="upper left", title="Building Type")
plt.title("Relationship between Area and Price in California Real Estate By Donald Programmer", fontsize=14, fontweight='bold')
plt.xlabel('Area (ft.)', fontweight='bold')
plt.ylabel('Price in (000)$')
plt.savefig('scatter.png')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x=data_real_estate_scatter['Area (ft.)'], 
            y=data_real_estate_scatter['Price'], 
            hue=data_real_estate_scatter['Building Type'],
            palette=['black', 'purple', 'darkblue', 'red', 'green'],    
            s=100
           )
plt.title("Relationship between Area and Price in California Real Estate By Donald Programmer", fontsize=14, fontweight='bold')
plt.xlabel('Area (ft.)', fontweight='bold')
plt.ylabel('Price in (000)$')
plt.savefig('scatter2.png')
plt.show()
```

# Linear regraion 

```python
data_ad_exp = pd.read_csv("scatter_plot_ii.csv")
```

```python
data_ad_exp
```

```python
sns.set(rc={'figure.figsize':(9, 6)})
sns.regplot(x="Budget", 
            y="Sales",
            data=data_ad_exp,
            scatter_kws={"color":"k"},
            line_kws={"color":"red"},)

plt.xlabel('Ad Expenditure in (000 $)')
plt.ylabel('Sale in (000) Unit')
plt.title('Effect of Ad Expenditure on Sale by Donald Programmer', fontsize=14, fontweight='bold')
plt.savefig('reg1.png')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9, 6)})
sns.lmplot(x="Budget", 
            y="Sales",
            data=data_ad_exp,
            scatter_kws={"color":"k"},
            line_kws={"color":"red"},
            height=10)

plt.xlabel('Ad Expenditure in (000 $)')
plt.ylabel('Sale in (000) Unit')
plt.title('Effect of Ad Expenditure on Sale by Donald Programmer', fontsize=14, fontweight='bold')
plt.savefig('reg.png')
plt.show()
```

# Bar-Line Chart

```python
data_paticipant_python = pd.read_csv('bar_line_chart_data.csv')
```

```python
data_paticipant_python
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.set_xlabel("Year", fontweight="bold")

ax.bar(data_paticipant_python['Year'],
        data_paticipant_python['Participants'], 
        color="k")

ax.set_ylabel("Number of participants", fontweight="bold")
ax.tick_params(axis="y",
               width=2,
              labelsize="large")
ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax=1.0))
ax1.plot(data_paticipant_python['Year'],
        data_paticipant_python['Python Users'],
        color="#b60000",
        marker="D")
ax1.set_ylabel("Python Users", fontweight="bold", color="#b60000")
ax1.tick_params(axis="y",
               width=2,
                color="#b60000",
              labelsize="large")

ax.set_title("Survey on Python Users From 2012 to 2019 by Donald Programmer", fontsize=14, fontweight='bold')
fig.savefig('python.png')
plt.show()
```
