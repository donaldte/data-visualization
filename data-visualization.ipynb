# Data Visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

# Bar chart

```python
chart_data_cars = pd.read_csv('bar_chart_data.csv')
chart_data_cars
```

```python
plt.figure(figsize=(9, 6))

plt.bar(chart_data_cars['Brand'], chart_data_cars['Cars Listings'], color=['r', 'g', 'b', 'w', "y", "m","c"])

plt.xticks(rotation=45, fontsize=13)

plt.yticks(fontsize=13)

plt.title("Cars Listing by Brand", fontsize=16, fontweight="bold")

plt.ylabel("Numbers of  Listing", fontsize=13)

plt.savefig("Used Cars Bar.pdf")

plt.show()
```

# Pie chart 

```python
pie_data_fuel = pd.read_csv('pie_chart_data.csv')
```

```python
pie_data_fuel
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(10, 8))

plt.pie(pie_data_fuel['Number of Cars'], labels=pie_data_fuel['Engine Fuel Type'].values, 
       autopct='%.2f%%',
       textprops = {
           'size': 'x-large',
           'fontweight': 'bold',
           'rotation': 30,
           'color': 'w'
       })

plt.legend()
plt.title(" Number of cars by Engine Fuel", fontweight="bold", fontsize=16)
plt.show()
```

# STACK CHART

```python
stack_fuel_data = pd.read_csv("stacked_area_chart_data.csv")
```

```python
stack_fuel_data
```

```python
plt.figure(figsize=(12, 6))
sns.set_style('white')
colors = ['#011638', '#7e2987', '#ef2026']
labels = ['Gas', 'Petrol', "Diesel"]
plt.stackplot(stack_fuel_data['Year'], stack_fuel_data['Gas'], stack_fuel_data['Petrol'], stack_fuel_data['Diesel'],
             colors=colors,
             edgecolor='none')

plt.xticks(stack_fuel_data['Year'],  rotation=45)
plt.ylabel("Number of Cars", fontsize=12)
plt.legend(labels=labels, loc='upper left')
plt.title("Popularity of engine fuel from 1982 to 2016", fontsize=16, fontweight='bold')
sns.despine()
plt.show()
```
