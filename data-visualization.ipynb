# Data Visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
chart_data_cars = pd.read_csv('bar_chart_data.csv')
chart_data_cars
```

```python
plt.figure(figsize=(9, 6))

plt.bar(chart_data_cars['Brand'], chart_data_cars['Cars Listings'], color=['r', 'g', 'b', 'w', "y", "m","c"])

plt.xticks(rotation=45, fontsize=13)

plt.yticks(fontsize=13)

plt.title("Cars Listing by Brand", fontsize=16, fontweight="bold")

plt.ylabel("Numbers of  Listing", fontsize=13)

plt.savefig("Used Cars Bar.pdf")

plt.show()
```
